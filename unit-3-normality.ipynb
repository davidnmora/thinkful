```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

NUM_SAMPLES = 100
```

```python
def hist_with_lines(dist, axes_constraints=False):
    plt.hist(dist)
    # Add a vertical line at the mean.
    plt.axvline(dist.mean(), color='b', linestyle='solid', linewidth=2)
    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(dist.mean() + dist.std(), color='b', linestyle='dashed', linewidth=2)
    #Add a vertical line at one standard deviation below the mean.
    plt.axvline(dist.mean() - dist.std(), color='b', linestyle='dashed', linewidth=2)
    
    if axes_constraints:
        plt.xlim(axes_constraints)
    plt.show()
```

# PART 1: Explorating six distributions

### 1. Normal

### 2. Laplace (probability *density* function)
In it's perfect symmetry, mean + std work great.

```python
hist_with_lines(np.random.laplace(loc=0, scale=1, size=NUM_SAMPLES))
```

### 3. Poisson: The probably`k` events occur within  `lam` interval - Probability *Mass* Function
At huge values of `size`, the Poisson approximates normal. Mean is somewhat meaningful in general, but `std` falls short because nearly every case of Poisson is *non-symmetric*.

```python
hist_with_lines(np.random.poisson(lam=5, size=NUM_SAMPLES))
```

### 4. Binomial: n trials with p probability of success 
Personally, the simple concept  behind this distribution means knowing `p` is perhaps the most descriptive value. That alone to me is more precise than having both mean and std, since it's more derivative.

```python
hist_with_lines(np.random.binomial(n=10, p=0.2, size=NUM_SAMPLES))
```

### 5. Bernoulli: special case of the binomial where n = 1
Again, just `p` here is most descriptive and clear.

```python
hist_with_lines(np.random.binomial(n=1, p=0.2, size=NUM_SAMPLES))
```

### 6. Logistic: continous probability function similar to normal, but with a larger tale
Here, it's similarity to `normal` distributions makes mean and std an acceptable way to describe it.

```python
hist_with_lines(np.random.logistic(loc=0, scale=1, size=100000))
```

# PART 2: Combining normal distributions
__Evaluation__: Fascinatingly, the summed normal distributions produce a *new normal distribution* whose `mean` and `std` is a sum of the individual means and stds.

```python
NUM_SAMPLES = 1000
std1 = 0.5
normal_1 = np.random.normal(loc=5, scale=std1, size=NUM_SAMPLES)
std2 = 1
normal_2 = np.random.normal(loc=10, scale=std2, size=NUM_SAMPLES)

normal_3 = normal_1 + normal_2

axes_constraints = (0,20)
hist_with_lines(normal_1, axes_constraints=axes_constraints)
hist_with_lines(normal_2, axes_constraints=axes_constraints)
hist_with_lines(normal_3, axes_constraints=axes_constraints)
```
